```python
# Importar os pacotes
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager #gestão de driver do G.Chorme
import pandas as pd
```

```python
# Instalando o Driver do Google Chrome
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
```

```python
# Requisição Web
url_site = 'https://www.fundamentus.com.br/resultado.php'
driver.get(url_site) # .get: acessa o link da variável url_site
```

```python
# Copiar o Full xpath via html da página
# Atribuir o caminho do dataframe
local_tabela = '/html/body/div[1]/div[2]/table'
```

```python
# Importar o dataframe
elemento =driver.find_element("xpath", local_tabela)
html_tabela = elemento.get_attribute("outerHTML")
html_tabela #print do dataframe
```

```python
# Lendo e formatando o dataframe 
tabela = pd.read_html(str(html_tabela), thousands = ".", decimal = ",")[0] #[0]: Primeiro dataframe no html
tabela #Print do dataframe
```

```python
# Fechar o navegador
driver.quit()
```
